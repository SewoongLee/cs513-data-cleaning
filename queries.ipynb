```python
con = sqlite3.connect('data.db')

cur = con.cursor()
cur.execute('''
create table Menu (
id integer primary key,
name text,
sponsor text,
event text,
venue text,
place text,
physical_description text,
occasion text,
notes text,
call_number text,
keywords text,
language text,
date text,
location text,
location_type text,
currency text,
currency_symbol text,
status text,
page_count integer,
dish_count integer
)
''')

cur.execute('''
create table Dish (
id integer primary key, 
name text,
description text,
times_appeared integer,
first_appeared integer,
last_appeared integer,
lowest_price real,
highest_price real
)
''')

cur.execute('''
create table MenuItem (
id integer primary key,
menu_page_id integer,
price integer,
high_price integer,
dish_id integer,
created_at text,
updated_at text,
xpos real,
ypos real
)
''')


cur.execute('''
create table MenuPage (
id integer primary key,
menu_id integer,
page_number integer,
image_id integer,
full_height real,
full_width real,
uuid text
)
''')
```

```python
# menu_df = pd.read_csv('data/original/Menu.csv')

cur.execute('begin')
try:
    for row in menu_df.to_dict(orient='records'):
        cur.execute('insert into Menu values (:id,:name,:sponsor,:event,:venue,:place,:physical_description,:occasion,:notes,:call_number,:keywords,:language,:date,:location,:location_type,:currency,:currency_symbol,:status,:page_count,:dish_count)', row)
    cur.execute('commit')
except:
    cur.execute('rollback')
    raise
```

```python
# df = pd.read_csv('data/original/Dish.csv')

cur.execute('begin')
try:
    for row in dish_df.to_dict(orient='records'):
        cur.execute('insert into Dish values (:id,:name,:description,:times_appeared,:first_appeared,:last_appeared,:lowest_price,:highest_price)', row)
    cur.execute('commit')
except:
    cur.execute('rollback')
    raise 
```

```python
menu_item_df = pd.read_csv('data/original/MenuItem.csv')

cur.execute('begin')
try:
    for row in menu_item_df.to_dict(orient='records'):
        cur.execute('insert into MenuItem values (:id,:menu_page_id,:price,:high_price,:dish_id,:created_at,:updated_at,:xpos,:ypos)', row)
    cur.execute('commit')
except:
    cur.execute('rollback')
    raise 
```

```python
# menu_page_df = pd.read_csv('data/original/MenuPage.csv')

cur.execute('begin')
try:
    for row in menu_page_df.to_dict(orient='records'):
        cur.execute('insert into MenuPage values (:id,:menu_id,:page_number,:image_id,:full_height,:full_width,:uuid)', row)
    cur.execute('commit')
except:
    cur.execute('rollback')
    raise 
```

```python
# find MenuPage entries that don't have parent Menu 
df = pd.read_sql_query('''
select count(distinct MenuPage.menu_id)
from MenuPage
''', con)

print(df)
# => 19816

df = pd.read_sql_query('''
select count(distinct MenuPage.menu_id)
from MenuPage
left join Menu
on MenuPage.menu_id = Menu.id
where Menu.id is null
''', con)

print(df)
# => 2271
```

```python
# Use SQL to fetch MenuPage entries that have related Menu entries
menu_page_df = pd.read_sql_query('''
select MenuPage.*
from MenuPage
left join Menu
on MenuPage.menu_id = Menu.id
where Menu.id is not null
''', con)
```

```python
# find MenuItem entries that don't have parent MenuPage
df = pd.read_sql_query('''
select count(distinct MenuItem.menu_page_id)
from MenuItem
''', con)

print(df)
# => 19816

df = pd.read_sql_query('''
select count(distinct MenuItem.menu_page_id)
from MenuItem
left join MenuPage
on MenuItem.menu_page_id = MenuPage.id
where MenuPage.id is null
''', con)

print(df)
# => 0 (no need for cleaning)
```

```python
# find MenuItem entries that don't have parent Dish 
df = pd.read_sql_query('''
select count(distinct MenuItem.dish_id)
from MenuItem
''', con)

print(df)
# => 19816

df = pd.read_sql_query('''
select count(distinct MenuItem.dish_id)
from MenuItem
left join Dish
on MenuItem.dish_id = Dish.id
where Dish.id is null
''', con)

print(df)
# => 3
```

```python
# Use SQL to fetch MenuPage entries that have related Menu entries
menu_item_df = pd.read_sql_query('''
select MenuItem.*
from MenuItem
left join Dish
on MenuItem.dish_id = Dish.id
where Dish.id is not null
''', con)
```
